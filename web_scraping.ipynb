```python
def append_df_to_excel(df, excel_path):
    df_excel = pd.read_excel(excel_path)
    result = pd.concat([df_excel, df], ignore_index=True)
    result.to_excel(excel_path, index=False)

def strip(x):
    if x !=None:
        return   str(x.text.strip())
    else:
        return str('')
```
